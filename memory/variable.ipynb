````python
# 定义全局变量
global_var = 42

def example_function():
    # 定义局部变量
    local_var = "Hello, World!"
    
    # 获取局部命名空间
    local_namespace = locals()
    print("Local Namespace:", local_namespace)
    
    # 获取局部变量名和对象的内存地址
    local_var_address = id(local_var)
    print(f"Local variable 'local_var' address: {local_var_address}")
    
    # 获取局部命名空间字典的内存地址
    local_namespace_address = id(local_namespace)
    print(f"Local namespace dictionary address: {local_namespace_address}")

# 调用函数
example_function()

# 获取全局命名空间
global_namespace = globals()
print("Global Namespace:", global_namespace)

# 获取全局变量名和对象的内存地址
global_var_address = id(global_var)
print(f"Global variable 'global_var' address: {global_var_address}")

# 获取全局命名空间字典的内存地址
global_namespace_address = id(global_namespace)
print(f"Global namespace dictionary address: {global_namespace_address}")
````

了解 Python 中的变量以及它们在计算机组成原理、操作系统和内存管理中的工作原理，可以帮助你更深入地理解编程和计算机科学的基础。以下是一些关键概念和它们的解释：

### Python 中的变量

1. **变量的定义**：
   - 在 Python 中，变量是一个名称，用于引用内存中的一个值。变量不需要显式声明类型，Python 是动态类型语言。
   - 例如：`x = 10`，这里 `x` 是一个变量，引用整数值 `10`。

2. **变量的类型**：
   - Python 支持多种数据类型，包括整数、浮点数、字符串、列表、字典等。
   - 变量的类型可以在运行时改变，例如：`x = 10` 然后 `x = "Hello"`。

3. **变量的作用域**：
   - 变量的作用域决定了变量的可见性和生命周期。Python 中有局部变量、全局变量和非局部变量。

### 计算机组成原理

1. **内存和存储**：
   - 内存（RAM）是计算机中用于存储临时数据的硬件。变量的值存储在内存中。
   - 存储器分为主存（RAM）和辅助存储（硬盘、SSD）。

2. **CPU 和寄存器**：
   - CPU 是计算机的中央处理单元，负责执行指令。寄存器是 CPU 内部的高速存储器，用于临时存储数据和指令。
   - 变量的值在计算时可能会被加载到寄存器中进行操作。

### 操作系统

1. **内存管理**：
   - 操作系统负责管理内存，包括内存分配和释放。Python 的内存管理由操作系统和 Python 解释器共同完成。
   - 操作系统使用虚拟内存技术，将物理内存和磁盘存储结合起来，为每个进程提供独立的地址空间。

2. **进程和线程**：
   - 进程是操作系统中运行的程序实例，每个进程有自己的内存空间。线程是进程中的执行单元，共享进程的内存空间。
   - Python 使用全局解释器锁（GIL）来管理多线程，确保同一时间只有一个线程执行 Python 字节码。

### 内存管理

1. **引用计数**：
   - Python 使用引用计数来管理内存。每个对象都有一个引用计数器，记录有多少引用指向该对象。当引用计数为零时，内存被释放。
   - 例如：`x = 10` 创建一个整数对象 `10`，并将 `x` 指向它。`y = x` 增加引用计数，当 `x` 和 `y` 都不再引用 `10` 时，内存被释放。

2. **垃圾回收**：
   - 除了引用计数，Python 还使用垃圾回收机制来处理循环引用。垃圾回收器会定期检查对象，回收不再使用的内存。
   - 例如：两个对象相互引用，但没有其他引用指向它们，垃圾回收器会检测到这种情况并释放内存。

### 示例代码

以下是一个简单的示例，展示了变量、引用计数和垃圾回收：

```python
import sys

# 创建变量并引用对象
x = [1, 2, 3]
y = x

# 查看引用计数
print(sys.getrefcount(x))  # 输出: 3 (x, y, getrefcount 的参数)

# 删除引用
del y
print(sys.getrefcount(x))  # 输出: 2 (x, getrefcount 的参数)

# 循环引用示例
a = []
b = [a]
a.append(b)

# 手动触发垃圾回收
import gc
gc.collect()
```

### 总结

- **变量**：在 Python 中，变量是对内存中值的引用，支持动态类型和不同的作用域。
- **计算机组成原理**：内存和 CPU 是变量存储和操作的基础。
- **操作系统**：负责内存管理和进程调度，Python 的内存管理依赖于操作系统。
- **内存管理**：Python 使用引用计数和垃圾回收来管理内存，确保有效使用和释放内存。

通过理解这些概念，你可以更深入地理解 Python 编程和计算机科学的基础。
