给你一份『词汇表』（字符串数组） words 和一张『字母表』（字符串） chars。

假如你可以用 chars 中的『字母』（字符）拼写出 words 中的某个『单词』（字符串），那么我们就认为你掌握了这个单词。

注意：每次拼写（指拼写词汇表中的一个单词）时，chars 中的每个字母都只能用一次。

返回词汇表 words 中你掌握的所有单词的 长度之和。

 

示例 1：

输入：words = ["cat","bt","hat","tree"], chars = "atach"
输出：6
解释： 
可以形成字符串 "cat" 和 "hat"，所以答案是 3 + 3 = 6。
示例 2：

输入：words = ["hello","world","leetcode"], chars = "welldonehoneyr"
输出：10
解释：
可以形成字符串 "hello" 和 "world"，所以答案是 5 + 5 = 10。
 

提示：

1 <= words.length <= 1000
1 <= words[i].length, chars.length <= 100
所有字符串中都仅包含小写英文字母

````python
class Solution:
    def countCharacters(self, words: list[str], chars: str) -> int:
        count = 0
        for word in words:
            word_len = len(word)
            for c in word:
                if c in chars:
                    word_len -= 1
            if word_len == 0:
                count += len(word)
        return count
    
s = Solution()
# 测试
print(s.countCharacters(["cat", "bt", "hat", "tree"], "atach"))  # 输出: 6
print(s.countCharacters(["hello", "world", "leetcode"], "welldonehoneyr"))  # 输出: 10
````

### BST (Binary Search Tree) 二叉搜索树

1. **节点结构**：每个节点包含一个键值（key），以及指向左子节点和右子节点的指针（或引用）。
2. **左子树性质**：对于每个节点，其左子树中的所有节点的键值都小于该节点的键值。
3. **右子树性质**：对于每个节点，其右子树中的所有节点的键值都大于该节点的键值。
4. **递归性质**：每个节点的左子树和右子树也都是二叉搜索树。

#### 二叉搜索树的基本操作

1. **插入（Insert）**：
   - 从根节点开始，比较新插入的键值与当前节点的键值。
   - 如果新键值小于当前节点的键值，则移动到左子节点；如果大于，则移动到右子节点。
   - 重复上述步骤，直到找到一个空位置，将新节点插入。

2. **搜索（Search）**：
   - 从根节点开始，比较要搜索的键值与当前节点的键值。
   - 如果键值相等，则找到该节点。
   - 如果键值小于当前节点的键值，则移动到左子节点；如果大于，则移动到右子节点。
   - 重复上述步骤，直到找到该节点或到达空节点（表示未找到）。

3. **删除（Delete）**：
   - 找到要删除的节点。
   - 如果该节点没有子节点，直接删除。
   - 如果该节点只有一个子节点，用其子节点替代它。
   - 如果该节点有两个子节点，找到其右子树中的最小节点（或左子树中的最大节点），用该节点替代要删除的节点，然后删除该最小节点（或最大节点）。

#### 二叉搜索树的优点

- **高效的查找、插入和删除操作**：在平均情况下，这些操作的时间复杂度为 \(O(\log n)\)，其中 \(n\) 是树中的节点数。
- **有序性**：中序遍历二叉搜索树可以得到一个有序的序列。

#### 二叉搜索树的缺点

- **不平衡性**：在最坏情况下（例如插入的节点是有序的），二叉搜索树会退化成链表，导致查找、插入和删除操作的时间复杂度变为 \(O(n)\)。

#### 示例代码

以下是一个简单的二叉搜索树的 Python 实现，包括插入、搜索和删除操作：

```python
class TreeNode:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

class BinarySearchTree:
    def __init__(self, arr=None):
        self.root = None
        if arr:
            for key in arr:
                self.insert(key)

    def insert(self, key):
        if self.root is None:
            self.root = TreeNode(key)
        else:
            self._insert(self.root, key)

    def _insert(self, node, key):
        if key < node.val:
            if node.left is None:
                node.left = TreeNode(key)
            else:
                self._insert(node.left, key)
        else:
            if node.right is None:
                node.right = TreeNode(key)
            else:
                self._insert(node.right, key)

    def search(self, key):
        return self._search(self.root, key)

    def _search(self, node, key):
        if node is None or node.val == key:
            return node
        if key < node.val:
            return self._search(node.left, key)
        else:
            return self._search(node.right, key)

    def delete(self, key):
        self.root = self._delete(self.root, key)

    def _delete(self, node, key):
        if node is None:
            return node
        if key < node.val:
            node.left = self._delete(node.left, key)
        elif key > node.val:
            node.right = self._delete(node.right, key)
        else:
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            temp = self._min_value_node(node.right)
            node.val = temp.val
            node.right = self._delete(node.right, temp.val)
        return node

    def _min_value_node(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current

    def inorder_traversal(self):
        return self._inorder_traversal(self.root, [])

    def _inorder_traversal(self, node, result):
        if node:
            self._inorder_traversal(node.left, result)
            result.append(node.val)
            self._inorder_traversal(node.right, result)
        return result

# 测试
keys = [9, 4, 7, 17, 10, 1, 25, 5]
bst = BinarySearchTree(keys)

print("Inorder traversal after insertion:", bst.inorder_traversal())

# 搜索节点
search_key = 10
found_node = bst.search(search_key)
if found_node:
    print(f"Node with key {search_key} found.")
else:
    print(f"Node with key {search_key} not found.")

# 删除节点
delete_key = 7
bst.delete(delete_key)
print(f"Inorder traversal after deleting {delete_key}:", bst.inorder_traversal())
```

#### 不同的BST (动态规划)

````python
def numTrees(n):
    # 初始化G数组，G[i]表示i个节点的二叉搜索树的种数
    G = [0] * (n + 1)
    G[0], G[1] = 1, 1

    # 计算G(2)到G(n)
    for i in range(2, n + 1):
        for j in range(1, i + 1):
            G[i] += G[j - 1] * G[i - j]

    return G[n]

# 测试
print(numTrees(3))  # 输出: 5
print(numTrees(1))  # 输出: 1
````

````python
class SequentialList:
    def __init__(self):
        self.data = []

    def insert(self, index, value):
        if index < 0 or index > len(self.data):
            raise IndexError("Index out of bounds")
        self.data.insert(index, value)

    def delete(self, index):
        if index < 0 or index >= len(self.data):
            raise IndexError("Index out of bounds")
        return self.data.pop(index)

    def find(self, value):
        try:
            return self.data.index(value)
        except ValueError:
            return -1

    def update(self, index, value):
        if index < 0 or index >= len(self.data):
            raise IndexError("Index out of bounds")
        self.data[index] = value

    def traverse(self):
        for item in self.data:
            print(item, end=" ")
        print()

# 使用顺序表
seq_list = SequentialList()

# 插入元素
seq_list.insert(0, 10)
seq_list.insert(1, 20)
seq_list.insert(2, 30)
seq_list.traverse()  # 输出: 10 20 30

# 删除元素
seq_list.delete(1)
seq_list.traverse()  # 输出: 10 30

# 查找元素
index = seq_list.find(30)
print(index)  # 输出: 1

# 更新元素
seq_list.update(1, 40)
seq_list.traverse()  # 输出: 10 40
````

### Linked List

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        
    def find_value(self, index):
        current = self.head
        for _ in range(index):
            if not current:
                raise IndexError("Index out of bounds")
            current = current.next
        return current
    
    def find_index(self, data):
        current = self.head
        index = 0
        while current:
            if current.data == data:
                return index
            current = current.next
            index += 1
        return -1

    def print_list(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# 使用链表
ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.print_list()  # 输出: 1 -> 2 -> 3 -> None
print(id(ll.find_value(0).data))  # 输出: 2
print(id(ll.find_value(1).data))  # 输出: 2
print(id(ll.find_value(2).data))  # 输出: 2
````

```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0

    while low <= high:
        mid = (high + low) // 2
        if arr[mid] < x:
            low = mid + 1
        elif arr[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1
```

线性表（Linear List）是一种数据结构，用于存储具有线性关系的数据元素。线性表可以分为两种主要类型：顺序表（Sequential List）和链表（Linked List）。

### 线性表的类型

1. **顺序表（Sequential List）**：使用连续的存储空间存储数据元素，通常用数组实现。
2. **链表（Linked List）**：使用节点（Node）和指针（Pointer）存储数据元素，节点通过指针连接。

### 顺序表的操作

顺序表的常见操作包括：

1. **初始化**：创建一个空的顺序表。
2. **插入**：在指定位置插入一个元素。
3. **删除**：删除指定位置的元素。
4. **查找**：查找指定元素的位置。
5. **更新**：更新指定位置的元素。
6. **遍历**：遍历顺序表中的所有元素。

### 示例代码

以下是使用 Python 列表实现顺序表及其常见操作的示例代码：

```python

    def __init__(self):
        self.data = []

        if index < 0 or index > len(self.data):
```

### 解释

1. **初始化**：
   - `__init__` 方法创建一个空的顺序表，使用 Python 列表 `self.data` 存储数据元素。

2. **插入**：
   - `insert` 方法在指定位置插入一个元素，使用 Python 列表的 `insert` 方法。

3. **删除**：
   - `delete` 方法删除指定位置的元素，使用 Python 列表的 `pop` 方法。

4. **查找**：
   - `find` 方法查找指定元素的位置，使用 Python 列表的 `index` 方法。如果元素不存在，返回 `-1`。

5. **更新**：
   - `update` 方法更新指定位置的元素，直接修改 Python 列表中的元素。

6. **遍历**：
   - `traverse` 方法遍历顺序表中的所有元素，使用 `for` 循环打印每个元素。
