给你一个整数 n ，求恰由 n 个节点组成且节点值从 1 到 n 互不相同的 二叉搜索树 有多少种？返回满足题意的二叉搜索树的种数。

 

示例 1：


输入：n = 3
输出：5
示例 2：

输入：n = 1
输出：1
 

提示：

1 <= n <= 19
这个题目怎么做，为啥我没有思路

这个问题实际上是一个经典的动态规划问题，通常称为“不同的二叉搜索树”问题。它可以通过计算卡塔兰数（Catalan Number）来解决。

### 思路

1. **定义状态**：
   - 设 `G(n)` 表示 `n` 个节点的二叉搜索树的种数。
   - 设 `F(i, n)` 表示以 `i` 为根节点的二叉搜索树的种数，其中 `1 <= i <= n`。

2. **状态转移方程**：
   - `G(n) = F(1, n) + F(2, n) + ... + F(n, n)`
   - 对于 `F(i, n)`，以 `i` 为根节点，左子树有 `i-1` 个节点，右子树有 `n-i` 个节点。
   - 因此，`F(i, n) = G(i-1) * G(n-i)`

3. **初始条件**：
   - `G(0) = 1`，空树只有一种。
   - `G(1) = 1`，只有一个节点的树也只有一种。

4. **计算过程**：
   - 使用动态规划，从 `G(0)` 计算到 `G(n)`。

### 代码实现

以下是 Python 的实现代码：

```python
def numTrees(n):
    # 初始化G数组，G[i]表示i个节点的二叉搜索树的种数
    G = [0] * (n + 1)
    G[0], G[1] = 1, 1

    # 计算G(2)到G(n)
    for i in range(2, n + 1):
        for j in range(1, i + 1):
            G[i] += G[j - 1] * G[i - j]

    return G[n]

# 测试
print(numTrees(3))  # 输出: 5
print(numTrees(1))  # 输出: 1
```

### 解释

1. **初始化**：
   - `G[0] = 1` 和 `G[1] = 1`，表示空树和只有一个节点的树的种数。

2. **动态规划**：
   - 外层循环从 `2` 到 `n`，表示计算 `G(2)` 到 `G(n)`。
   - 内层循环从 `1` 到 `i`，表示以 `j` 为根节点的情况。
   - `G[i] += G[j - 1] * G[i - j]`，累加以 `j` 为根节点的所有情况。

3. **返回结果**：
   - 最终返回 `G[n]`，即 `n` 个节点的二叉搜索树的种数。

这种方法的时间复杂度是 \(O(n^2)\)，空间复杂度是 \(O(n)\)，可以在合理的时间内计算出结果。
